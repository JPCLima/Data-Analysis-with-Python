# Plotting and Visualization 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creating a array of numbers with linspace(start, end, step)
x = np.linspace(0, 10, 100) 
y = np.sin(x)
plt.plot(x, y)
plt.show
```

```python
# Labels, title and legend
x = np.linspace(0, 10, 100) 
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sin signal')
plt.show
```

```python
# Mutiple plots in one figure
x = np.linspace(0, 10, 100) 
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y, label='sin(x)')
plt.plot(x, z, label='cos(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sin signal')
plt.show
```

## Scatterplot

```python
# Creating a random numbers using randn
data = np.random.randn(100,2, 5)
x = data[:,0]
y = data[:,1]
# Scatter Plot
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show
plt.show
```

## Histogram

```python
# Histogram plot of uniform distribution
num = np.random.random(1000)
plt.hist(num)
plt.show
```

```python
# Histogram plot of normal distribution
num = np.random.randn(1000)

# Control the number of buckets using bins
plt.hist(num, bins=100)
plt.show
```

## Figures and subplots

```python
# Creates two subplots - subplots(rows, cols)
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.scatter(x, y)
ax1.set_title('Scatter Plot')
ax2.hist(num)
ax2.set_title('Histogram Plot')
```

## Color, Markers, Line Style and Ticks

```python
# Creating random data
data = np.random.random(50)
# Plot data
plt.plot(data, color='r', linestyle='--', marker='o', markerfacecolor='blue')
plt.xticks([0, 10, 20, 30, 40, 50, 60])
```

## Annotation 

```python
x = np.linspace(-100,100)
parab = x**2
plt.ylim(-4000,11000)

arrow = dict(
    facecolor="black", width=1,
    headwidth=5, shrink=0.1)

plt.annotate(
    "min", xy=(25,0),
    xytext=(50, 0),
    arrowprops=arrow)

plt.plot(parab)

```
